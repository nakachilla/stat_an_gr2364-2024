$\vec{X}=(X_1,...,X_n)^T ,\vec{\xi}=\mathcal{N}(\mu, \Sigma)$

$ A=(a_{ij})_{1\leq i,j\leq n},b=(b_1,...,b_n)^T  $

$ \vec{\eta}=A\vec{\xi}+b\sim \mathcal{N}(A\mu+b,A \Sigma A^T) $

5. Найти условное распределение $ \xi $ при условии $ \eta $. Вычислить $ \mathbb{E}(\xi|\eta) $ и $ \mathbb{D}(\xi|\eta). $ 

Плотность распределения  $ \xi $ при условии $ \eta $:

$ \rho_{\xi|\eta=y_0}(x)=\frac{\rho_{\xi,\eta}(x,y_0)}{\rho_{\eta}(y_0)} $

$ \rho_{\eta}(y_0)=\int_{-\infty}^{+\infty}\rho_{\xi,\eta}(x,y)dx = \frac{\sqrt{2\pi}}{\sigma_{\eta}}exp\left(-\frac{1}{2}\cdot\left(\frac{y-a_{\eta}}{\sigma_{\eta}}\right)^2  \right) $

$$ \frac{\rho_{\xi,\eta}(x,y_0)}{\rho_{\eta}(y_0)}=\frac{\frac{5\sqrt{2}}{2\pi}\cdot \exp\left(-\frac{1}{2}\left( 6x^2-4xy_0+9y_0^2+8x+14y_0+11 \right)\right)}{\frac{5\sqrt{2\pi}}{\sqrt{3}}\exp\left(-\frac{1}{2}\cdot\left(\frac{y+1}{\frac{\sqrt{3}}{5}}\right)^2  \right)}   $$

Преобразовав дробь $ \frac{\rho_{\xi,\eta}(x,y_0)}{\rho_{\eta}(y_0)}  $, получим:

$$ \rho_{\xi|\eta=y_0}(x)= \frac{\sqrt{3\pi}}{\pi}exp\left(-\frac{1}{2}\cdot\left(\frac{x-(\frac{y_0-2}{3})}{\frac{1}{\sqrt{6}}}\right)^2  \right) $$

Тогда $ \mathbb{E}(\xi|\eta)=\frac{y_0-2}{3} $

$ \mathbb{D}(\xi|\eta)=\frac{1}{6} $
